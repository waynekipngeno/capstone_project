### Data Aggregation:
With naming and date harmonization complete, we can now work on aggregating the two dataframes. We need only a few columns from the api_dataframe, we can drop the other columns.

```python
# Keep only the specified columns in the api_data DataFrame
api_data_subset = api_data[['id', 'homeID','awayID', 'home_name', 'away_name', 'date', 'season', 'status', 'homeGoalCount', 'awayGoalCount',
                            'totalGoalCount', 'team_a_cards_num', 'team_b_cards_num', 'odds_ft_1', 'odds_ft_x',
                            'odds_ft_2', 'home_ppg', 'away_ppg', 'team_a_xg_prematch', 'team_b_xg_prematch',
                            'total_xg_prematch', ]]
```

```python
api_data_subset.head()
```

```python
# Define a list of features from the 'scraped_data' dataframe to include in the merged dataframe
scraped_data_feats = ['xG', 'xGA', 'npxG', 'npxGA', 'deep', 'deep_allowed', 'scored', 'missed', 
                      'xpts', 'npxGD', 'ppda.att', 'ppda.def', 'ppda_allowed.att', 'ppda_allowed.def']
```

```python
home_id = 150

# Convert the date string to a datetime object
date = pd.to_datetime('2016-08-13').date()
```

```python
# Sort the DataFrame by the 'date' column in ascending order
sorted_scraped_data = scraped_data.sort_values(by='date', ascending=False)
```

```python
result = sorted_scraped_data[(sorted_scraped_data['teamID'] == home_id) & (sorted_scraped_data['date'] < date)].head().describe().loc['mean']
```

```python
result
```

```python
result_dict = result.to_dict()
```

```python
result_dict
```

```python
# {key+ f'_home':value for key, value in filtered_result_dict[0].items() }
```

```python
# Create empty columns for home and away features
for feat in scraped_data_feats:
    api_data_subset.loc[:, f'{feat}_home'] = None
    api_data_subset.loc[:, f'{feat}_away'] = None

```

```python
api_data_subset.head()
```

```python
# Iterate through each row in api_data_subset
for index, row in api_data_subset.iterrows():
    home_id = row['homeID']
    game_date = row['date']
    
    # Retrieve stats from scraped_data dataframe
    result = sorted_scraped_data[(sorted_scraped_data['teamID'] == home_id) & (sorted_scraped_data['date'] < date)]
    # get mean for each statistics for the last  5 games before current match
    result_mean = result.head(15).describe().loc['mean']
    # Check if there are matching records
    if not result.empty:
        # Take the first matching record
        filtered_result_dict = result_mean.to_dict()
        
        # Filter out irrelevant columns from result
        filtered_result_dict = {key: value for key, value in filtered_result_dict.items() if key in scraped_data_feats}
        
        # Update values in the dataframe
        for key, value in filtered_result_dict.items():
            api_data_subset.at[index, f'{key}_home'] = value
```

```python
# Iterate through each row in api_data_subset
for index, row in api_data_subset.iterrows():
    away_id = row['awayID']
    game_date = row['date']

    
    # Retrieve stats from scraped_data dataframe
    result = sorted_scraped_data[(sorted_scraped_data['teamID'] == away_id) & (sorted_scraped_data['date'] < game_date)]
    # get mean for each statistics for the last  5 games before current match
    result_mean = result.head(15).describe().loc['mean']
       
    # Check if there are matching records
    if not result.empty:
        # Take the first matching record
        filtered_result_dict = result_mean.to_dict()
        
        # Filter out irrelevant columns from result
        filtered_result_dict = {key: value for key, value in filtered_result_dict.items() if key in scraped_data_feats}
        
        # Update values in the dataframe
        for key, value in filtered_result_dict.items():
            api_data_subset.at[index, f'{key}_away'] = value
```
